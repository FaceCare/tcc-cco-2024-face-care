```python
import boto3
import os
```

```python
boto3.setup_default_session(profile_name="faculdade")
s3 = boto3.client('s3')
```

```python
bucket_staged = 'tcc-dev-staged-bucket'
response = s3.list_objects_v2(Bucket=bucket_staged, Prefix='')

files_s3 = [item['Key'] for item in response['Contents']]
files_acne = [f for f in files_s3 if str(f).startswith('Acne/')]
files_no_acne = [f for f in files_s3 if str(f).startswith('SemAcne/')]

def download_list_files_s3(files: list, bucket_name: str, save_dir=''):
    if not os.path.exists(save_dir):
        os.mkdir(save_dir)
        
    for f in files:
        print(f'Downloading file={f}')
        s3.download_file(bucket_name, f, os.path.join(save_dir, os.path.split(f)[-1]))

download_list_files_s3(files_acne, bucket_staged, 'Acne')
download_list_files_s3(files_no_acne, bucket_staged, 'SemAcne')
```
